## Visualizing results from the Instance Segmentation Mechanical Turk task

Set `batch_csv_file` to the path of the csv file with batch results form a mturk task, and run all.

```python
import base64
import io
import numpy as np
from PIL import Image
import cv2
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import requests
import urllib.request
from IPython.core.display import display, HTML

pd.set_option('max_colwidth', 300)
matplotlib.rcParams['figure.figsize'] = (20, 7.0)

# SET THIS VARIABLE AND RUN ALL. This is the path to the csv file for batch results 
# form a mturk task.
batch_csv_file = 'mc_new/locobot/notebooks/turk_annotations/Batch_4275491_batch_results.csv'

display(HTML(
    """
    <style>
    .container { width:100% !important; }
    #notebook { letter-spacing: normal !important;; }
    .CodeMirror { font-family: monospace !important; }
    .cm-keyword { font-weight: bold !important; color: #008000 !important; }
    .cm-comment { font-style: italic !important; color: #408080 !important; }
    .cm-operator { font-weight: bold !important; color: #AA22FF !important; }
    .cm-number { color: #080 !important; }
    .cm-builtin { color: #008000 !important; }
    .cm-string { color: #BA2121 !important; }
    </style>
    """
))


def decodeAndImg(image_dat):
    """decodes byte-64 encoded image data.
    """
    image = base64.b64decode(str(image_dat))       
    img = Image.open(io.BytesIO(image))
    return img.convert('RGB')

def url_to_image(url):
    """Converts an image url to a PIL image.
    """
    resp = urllib.request.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return Image.fromarray(image)

def displayResults(filename):
    """visualizes an mturk batch.
    
    Visualizes each image, along with the masks drawn for each image. 
    Outputs turker ID that can be used to (manually) select turkers that perform well.
    Args:
        filename: batch results csv file from mturk
    """
    df = pd.read_csv(filename)
    
    for index, row in df.iterrows():
        print("WorkerID {} Time take in minutes {}".format(row['WorkerId'], row['WorkTimeInSeconds']/60))
        img = decodeAndImg(row['Answer.annotatedResult.labeledImage.pngImageData'])
        # Download the actual image 
        orig = url_to_image(row['Input.image_url'])
        x,y = img.size
        c = Image.blend(img, orig, 0.4)
        plt.figure(figsize=(10,10))
        plt.imshow(c)
        plt.show()

displayResults(batch_csv_file)
```
